### help()

Esta función se utiliza para mostrar la documentación de un objeto en Python.

```python
help(len)
```

### print()

```python
print("carlos")
```

```python
print(1)
```

```python
print(len)
```

```python
print(type(2))
```

### str(), int(), float()

```python
str(2)
```

```python
int('1')
```

```python
float("2312312312.321312")
```

### type()

Retorna el tipo del objeto que le proporcionamos como argumento.

```python
type(1)
```

```python
type(len)
```

```python
from os import MFD_ALLOW_SEALING
type(MFD_ALLOW_SEALING)
```

### id()

```python
help(id)
```

```python
var = "myobject"
```

```python
id(var)
```

Debemos tener en cuenta que el identificador hace referencia al objeto, en este caso al string, y no a la variable a la que esta asignado.

```python
id("myobject")
```

```python
id("myobject2")
```

### exec()

```python
help(exec)
```

```python
while True:
  exec(input("->> "))
```

```python
print(exec("1+2"))
```
