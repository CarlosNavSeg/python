````python
def crear_tablero (filas, columnas):
  tablero = [None]*filas
  for i in range(filas):
    tablero[i] = ["."]*columnas
  return tablero
````

````python
tablero_juego = crear_tablero(6,7)
def mostrar_tablero(tablero):
    """Muestra el tablero por pantalla."""
    # Sacamos por pantalla la cabecera
    for num in range(len(tablero[0])):
        print(num, end='  ')
    # Sacamos por pantalla el tablero
    for fila in tablero:
        print("")
        for casilla in fila:
            print(casilla, end="  ")
````

````python
def introducir_ficha(tablero, columna, color):
  ficha_type = ""
  if color == "rojo":
    ficha_type = "-"
  elif color == "azul":
    ficha_type = "x"
  else:
    return "El color no es válido"
  if(tablero_juego[len(tablero_juego)-1][columna] == "."):
    tablero_juego[len(tablero_juego)-1][columna] = ficha_type
  else:
    for i in range(0, len(tablero_juego)-1):
      if(tablero[len(tablero_juego)-2-i][columna] == "."):
        tablero_juego[len(tablero_juego)-2-i][columna] = ficha_type
        return tablero
  return tablero
````
